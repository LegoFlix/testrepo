## Filter Dataframe based on certain criteria

```python
# a filter for ConvertedComp that is greater than 70,000
high_salary = (df_1['ConvertedComp'] > 70000)

# Apply filter to the dataframe using .loc() and pass in the 
# high_salary filter:
df_1.loc[high_salary]
```

```python
# include multiple columns with the salary filter:
df_1.loc[high_salary, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
# Filter with multiple values:
# 1 - create a list of what needs filtering, eg, US, india, UK, germany
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
filter_countries = df_1['Country'].isin(countries)
df_1.loc[filter_countries]
```

```python
df_1.loc[filter_countries, 'Country']
```

```python
# filter for people who work with Python:
df_1['LanguageWorkedWith']

# There is a semi-colon separator, so use a string method 
# and see if python is in the string
```

```python
python1 = df_1['LanguageWorkedWith'].str.contains('Python', na=False)
# for languages worked with, does the string contain python
# and ignore any NaN
df_1.loc[python1, 'LanguageWorkedWith']
# string methods are usefulto replace text, split valuespython1
```
